### 1. Idea of the Light Clock

We imagine a simple clock made of two mirrors. A light pulse travels between them. Every time the light goes from the bottom mirror to the top mirror and back, the clock measures one tick.

If the distance between the mirrors is $d$, and the speed of light is $c$, the light travels vertically between the mirrors.

For a stationary observer, the path of the light is straight up and down.

```python
d = 5  # distance between mirrors
frames = 120

fig, ax = plt.subplots(figsize=(4,6))

ax.set_xlim(-1,1)
ax.set_ylim(-1,d+1)

ax.set_title("Light Clock (Stationary)")
ax.set_xlabel("horizontal position")
ax.set_ylabel("vertical position")

# mirrors
ax.scatter([0,0],[0,d], color="black")

ax.text(0.1,d,"Top mirror")
ax.text(0.1,0,"Bottom mirror")

light, = ax.plot([], [], 'ro', markersize=10)

def animate(frame):

    period = 60
    
    phase = frame % (2*period)

    if phase < period:
        y = (phase/period) * d
    else:
        y = d - ((phase-period)/period) * d

    light.set_data([0],[y])

    return light,

ani = FuncAnimation(fig, animate, frames=frames, interval=40)

HTML(ani.to_jshtml())
```
